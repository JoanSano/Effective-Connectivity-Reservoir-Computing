```python
import numpy as np
import matplotlib.pyplot as plt
#https://matplotlib.org/stable/gallery/mplot3d/lorenz_attractor.html

def lorenz(xyz, *, s=10, r=28, b=2.667):
    """
    Parameters
    ----------
    xyz : array-like, shape (3,)
       Point of interest in three-dimensional space.
    s, r, b : float
       Parameters defining the Lorenz attractor.

    Returns
    -------
    xyz_dot : array, shape (3,)
       Values of the Lorenz attractor's partial derivatives at *xyz*.
    """
    x, y, z = xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return np.array([x_dot, y_dot, z_dot])
```

```python
dt = 0.01
num_steps = 10000

xyzs = np.empty((num_steps + 1, 3))  # Need one more for the initial values
xyzs[0] = (0., 1., 1.05)  # Set initial values
# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point
for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt

fmt = "svg"
```

```python
# Plot
ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')

ax.plot(*xyzs.T, lw=1)
ax.set_xticks([]), ax.set_yticks([]), ax.set_zticks([])
ax.set_xlabel(r"$x_{t}$", fontsize=15)
ax.set_ylabel(r"$y_{t}$", fontsize=15)
ax.set_zlabel(r"$z_{t}$", fontsize=15)
ax.dist = 11
#plt.savefig("Oringinal."+fmt, format=fmt, dpi=700)
plt.show()
```

```python
# Plot
ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')

#ax.plot(*xyzs.T, lw=0.5)
ax.plot(xyzs[48:-2,0],xyzs[40:-10,0],xyzs[30:-20,0], lw=1)
ax.set_xticks([]), ax.set_yticks([]), ax.set_zticks([])
ax.set_xlabel(r"$x_{t}$", fontsize=15)
ax.set_ylabel(r"$x_{t-8}$", fontsize=15)
ax.set_zlabel(r"$x_{t-18}$", fontsize=15)
ax.dist = 11
#plt.savefig("Oringinal_shadow."+fmt, format=fmt, dpi=700)
plt.show()
```

```python
# Reverse variable
v = 1
reversed_xyzs = np.copy(xyzs)
reversed_xyzs[:,v-1] = xyzs[::-1,v-1]

# Plot
ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')
ax.plot(reversed_xyzs[48:-2,v-1],reversed_xyzs[40:-10,v-1],reversed_xyzs[30:-20,v-1], lw=1)
ax.set_xticks([]), ax.set_yticks([]), ax.set_zticks([])
ax.set_xlabel(r"$x_{-t}$", fontsize=15)
ax.set_ylabel(r"$x_{-(t-8)}$", fontsize=15)
ax.set_zlabel(r"$x_{-(t-18)}$", fontsize=15)
ax.dist = 11
#plt.savefig("X_reversed_shadow."+fmt, format=fmt, dpi=700)
plt.show()
```
